identity operators

```python
a = [1,2,3,4] #list__mutable type changes(delete,insert,change)
b = [1,2,3,4]
print(a == b)
```

```python
print(a is b)
```

```python
c = 12 #immutable datatype can't change
d = 12
print(c == d) #comparison operator
print(c  is d) #identity operator
```

```python
print("id of a:",id(a)) #memory address of a
print("id of b:",id(b)) #memory address of b
print("id of c:",id(c)) #memory address of c
print("id of d:",id(d)) #memory address of d
```
