```python
#Nested Loop -- A loop inside another loop
```

```python
for i in range(1,4):
  for j in range(1,4):
    print(f"{i},{j}")
```

```python
#use case

'''1. to work with multidimensional data
2. to find out possible combinations'''
```

```python
l1 = [1,2,3,4,5]
l2 = [1,2,3,4,5]
for element1 in l1:
  for element2 in l2:
    print(f"{element1},{element2}")
```

```python
str1 = 'jaipur'
str2 = 'delhi'
for char1 in str1:
  for char2 in str2:
    print(f"{char1},{char2}")
```

```python
t1 = (1,2,3,4,5)
t2 = (1,2,3,4,5)
for element1 in t1:
  for element2 in t2:
    print((element1)*(element2))
```

```python
#control statement
'''1. break
2. continue
3. pass'''
```

```python
#break
'''break the loop prematurely'''

for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
#continue
'''skip the current iteration and continue with the next iterartion'''

for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
#pass
'''do nothing
used for syntactical reasons only'''
for i in range(1,10):
  if i ==4:
    pass
  print(i)
```

```python
#else
'''executes after the loop is finished'''

for i in range(1,10):
  if i == 3:
   pass
  print(i)


else:
    print("Loop finished")
```

```python
#infinite loops
'''Loops that never end'''

while True:
  print("hello")
```

```python
#enumerate function
'''when the values are required along with their corresponding index'''

fruits = ['apple','banana','cherry']
for index, fruit in enumerate(fruits):
  print(index,fruit)
```

```python
#accessing keys and values

d1 = {'a':1,'b':2,'c':3}
for key,value in d1.items():
  print(key,value)
```
