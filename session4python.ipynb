```python
#list methods --- append,extend,insert,remove,pop,clear,sort,reverse,copy,count,index,len,max,min
```

```python
l1 = [1,2,3,4,'apple',3.4]
l1.append(10)   #adds a single element at the end of the list
print(l1)
```

```python
l1.extend([11,12,13])  #adds multiple elements at the end of the list
print(l1)
```

```python
#takes two parameters--1st-index 2nd-element
l1.insert(2,15) #inserts element 15 at index 2
print(l1)
```

```python
l1.remove(2) #removes the element
#if elements are repeated it removes only the first occurence of the element
print(l1)
```

```python
l2 = [1,2,2,3,4]
l2.remove(2)
print(l2)
```

```python
l1.pop(2) #removes element at index 2
#takes index as its parameter
print(l1)
```

```python
lst = [1,2,3,4,5.5,'apple','mango']
lst.append('a')
print("after appending\n",lst)
lst.extend([2,'b'])
print("after extending\n",lst)
lst.insert(2,'c')
print("after inserting\n",lst)
lst.remove(4)
print("after removing\n",lst)
lst.pop(5)
print("after popping\n",lst)
```

```python
print(l2)
l2.pop() #pops the last element
print(l2)
```

```python
num = [4,56,7,3,3,5]
num.sort() #sorts the list
print(num)
```

```python
l2.clear() #clears the list
#don't accept any argument
print(l2)
```

```python
lst.reverse() #reverses the list
print(lst)
```

```python
l4 = l1.copy() #copies the list
print(l4)
```

```python
letters = ['a','c','f','b','d','e']
letters.sort()
print(letters)
cletters = letters.copy()
print(cletters)
letters.pop()
print(letters)
print(cletters)
letters.reverse()
print(letters)
print(cletters)
letters.clear()
print(letters)
print(cletters)
```

```python
l7 = [1,2,5,4,3,4,4,4,4,4,]
l7.count(4) #counts the number of occurences of an element
```

```python
l7.index(4,3,9) #accept three parameters---- element(compulsary),starting index(optional),ending index(optional)
```

```python
len(l7) #returns the length of the list
```

```python
min(l7) #returns the minimum value of the list
```

```python
max(l7) #returns the maximum value of the list
```
